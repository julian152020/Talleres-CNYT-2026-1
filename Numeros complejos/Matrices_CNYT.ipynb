# 1. Adición de vectores complejos.

```python
n = int(input("Ingrese el tamaño del vector: "))

print("\nIngrese los valores del primer vector:")
v1 = []
for i in range(n):
    valor = complex(input(f"Elemento {i+1}: "))
    v1.append(valor)

print("\nIngrese los valores del segundo vector:")
v2 = []
for i in range(n):
    valor = complex(input(f"Elemento {i+1}: "))
    v2.append(valor)

v1 = np.array(v1, dtype=complex)
v2 = np.array(v2, dtype=complex)

resultado = v1 + v2

print("\nResultado:")
print(resultado)
```

# 10. Producto de dos matrices.

```python
f1 = int(input("Filas matriz A: "))
c1 = int(input("Columnas matriz A: "))

A = []
for i in range(f1):
    fila = []
    for j in range(c1):
        fila.append(complex(input(f"A[{i+1},{j+1}]: ")))
    A.append(fila)

A = np.array(A, dtype=complex)

f2 = int(input("\nFilas matriz B: "))
c2 = int(input("Columnas matriz B: "))

B = []
for i in range(f2):
    fila = []
    for j in range(c2):
        fila.append(complex(input(f"B[{i+1},{j+1}]: ")))
    B.append(fila)

B = np.array(B, dtype=complex)

if c1 != f2:
    print("\nError: dimensiones incompatibles.")
else:
    resultado = A @ B
    print("\nResultado:")
    print(resultado)
```

# 11. Accion de una matriz sobre un vector.

```python
f = int(input("Filas de la matriz: "))
c = int(input("Columnas de la matriz: "))

M = []
for i in range(f):
    fila = []
    for j in range(c):
        fila.append(complex(input(f"M[{i+1},{j+1}]: ")))
    M.append(fila)

M = np.array(M, dtype=complex)

print("\nVector (tamaño debe ser igual a columnas):")
v = []
for i in range(c):
    v.append(complex(input(f"v[{i+1}]: ")))

v = np.array(v, dtype=complex)

resultado = M @ v

print("\nResultado:")
print(resultado)
```

# 12. Producto interno de dos vectores.

```python
n = int(input("Tamaño del vector: "))

v1 = []
for i in range(n):
    v1.append(complex(input(f"v1[{i+1}]: ")))

v2 = []
for i in range(n):
    v2.append(complex(input(f"v2[{i+1}]: ")))

v1 = np.array(v1, dtype=complex)
v2 = np.array(v2, dtype=complex)

resultado = np.vdot(v1, v2)

print("\nResultado:")
print(resultado)
```

# 13. Norma de un vector.

```python
n = int(input("Tamaño del vector: "))

v = []
for i in range(n):
    v.append(complex(input(f"v[{i+1}]: ")))

v = np.array(v, dtype=complex)

resultado = np.linalg.norm(v)

print("\nResultado:")
print(resultado)
```

# 14. Distancia entre dos vectores.

```python
n = int(input("Tamaño del vector: "))

v1 = []
for i in range(n):
    v1.append(complex(input(f"v1[{i+1}]: ")))

v2 = []
for i in range(n):
    v2.append(complex(input(f"v2[{i+1}]: ")))

v1 = np.array(v1, dtype=complex)
v2 = np.array(v2, dtype=complex)

resultado = np.linalg.norm(v1 - v2)

print("\nResultado:")
print(resultado)
```

# 15. Valores y vectores propios de una matriz.

```python
n = int(input("Tamaño de la matriz cuadrada: "))

M = []
for i in range(n):
    fila = []
    for j in range(n):
        fila.append(complex(input(f"M[{i+1},{j+1}]: ")))
    M.append(fila)

M = np.array(M, dtype=complex)

valores, vectores = np.linalg.eig(M)

print("\nValores propios:")
print(valores)

print("\nVectores propios (columnas):")
print(vectores)
```

# 16. Revisar si una matriz es unitaria.

```python
n = int(input("Tamaño de la matriz cuadrada: "))

M = []
for i in range(n):
    fila = []
    for j in range(n):
        fila.append(complex(input(f"M[{i+1},{j+1}]: ")))
    M.append(fila)

M = np.array(M, dtype=complex)

resultado = np.allclose(M.conj().T @ M, np.eye(n))

print("\n¿Es unitaria?")
print(resultado)
```

# 17. Revisar si una matriz es hermitiana.

```python
n = int(input("Tamaño de la matriz cuadrada: "))

M = []
for i in range(n):
    fila = []
    for j in range(n):
        fila.append(complex(input(f"M[{i+1},{j+1}]: ")))
    M.append(fila)

M = np.array(M, dtype=complex)

resultado = np.allclose(M, M.conj().T)

print("\n¿Es Hermitiana?")
print(resultado)
```

# 18. Producto tensor de dos matrices/vectores.

```python
f1 = int(input("Filas matriz A: "))
c1 = int(input("Columnas matriz A: "))

A = []
for i in range(f1):
    fila = []
    for j in range(c1):
        fila.append(complex(input(f"A[{i+1},{j+1}]: ")))
    A.append(fila)

A = np.array(A, dtype=complex)

f2 = int(input("\nFilas matriz B: "))
c2 = int(input("Columnas matriz B: "))

B = []
for i in range(f2):
    fila = []
    for j in range(c2):
        fila.append(complex(input(f"B[{i+1},{j+1}]: ")))
    B.append(fila)

B = np.array(B, dtype=complex)

resultado = np.kron(A, B)

print("\nResultado:")
print(resultado)
```
